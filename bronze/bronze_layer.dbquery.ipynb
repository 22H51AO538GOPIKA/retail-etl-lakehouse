```sql
USE CATALOG `retail-dwh-project`;


USE SCHEMA bronze;


-- Drop all existing tables first to avoid location conflicts
DROP TABLE IF EXISTS bronze.customers_raw;
DROP TABLE IF EXISTS bronze.products_raw;
DROP TABLE IF EXISTS bronze.stores_raw;
DROP TABLE IF EXISTS bronze.sales_raw;


-- Now create all tables with correct paths
CREATE TABLE bronze.customers_raw
USING CSV
OPTIONS (
  path 's3://retail-etl-lakehouse-gopika/raw/customers/',
  header 'true',
  inferSchema 'true'
);


CREATE TABLE bronze.products_raw
USING CSV
OPTIONS (
  path 's3://retail-etl-lakehouse-gopika/raw/products/',
  header 'true',
  inferSchema 'true'
);


CREATE TABLE bronze.stores_raw
USING CSV
OPTIONS (
  path 's3://retail-etl-lakehouse-gopika/raw/stores/',
  header 'true',
  inferSchema 'true'
);


CREATE TABLE bronze.sales_raw
USING CSV
OPTIONS (
  path 's3://retail-etl-lakehouse-gopika/raw/sales/',
  header 'true',
  inferSchema 'true'
);
```
